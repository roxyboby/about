```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

```

```python
xx = np.linspace(-3,3,51)
X,Y = np.meshgrid(xx,xx)

f1 = 4
f2 = .8
p1 = .6
p2 = -.77

Z = np.sin(X*np.cos(Y))
gx,gy = np.gradient(Z)

plt.contourf(xx,xx,Z,40, cmap="Blues");
plt.quiver(xx,xx,gx,gy);
```

```python
xDomain = [-2,2]
# Option-1
resolution = .1
# x = np.arange(xDomain[0],xDomain[1] + resolution,resolution)


# Option-2
x = np.linspace(xDomain[0],xDomain[1],41)
y = x**2 + 3*x**3 -x**4 

# Plot
plt.plot (x,y, label = '$y=x^2 + 3*x^3 -x^4$')
plt.legend()
plt.grid()
plt.xlim(xDomain[0],xDomain[1])
plt.xlabel("x")
plt.ylabel('y = f(x)')
plt.show()
```

```python
# Create a symbolic variable
s_beta = sym.var('beta')

# Define the function
s_y = s_beta**2 + 3*s_beta**3 - s_beta**4
sym.plot(s_y, (s_beta, xDomain[0], xDomain[1]), xlabel="x", ylabel=None, title=f'$f(\\beta) = {sym.latex(s_y)}$')
```

```python
# Converting from Sympy to Numpy
fx = sym.lambdify(s_beta, s_y)
#fx(2)
yy = fx(x)

plt.plot (x,yy, label = '$y=x^2 + 3*x^3 -x^4$')
plt.legend()
plt.grid()
plt.xlim(xDomain[0],xDomain[1])
plt.xlabel("x")
plt.ylabel('y = f(x)')
plt.show()
```
